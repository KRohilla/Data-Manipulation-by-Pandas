Loading Data from CSV file to Dataframe

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/01.Data Cleaning and Preprocessing.csv')
# random 5 rows from the data
data.sample(5)
```

```python
# finding the unique values in each column
for i in data.columns:
  print(i)
  print(np.unique(data[i]))
```

```python
data = data.drop('Observation', axis=1)
```

```python
# starting 5 rows from the data
data.head()
# for last 5 rows from the data
# data.tail()
```

```python
# checking the datatype of our data
type(data)
```

```python
# shape of the data dataframe
data.shape
```

```python
# total number of duplicated rows present in the data
data.duplicated().sum()
```

```python
# dropping the duplicated rows
data = data.drop_duplicates()
```

```python
# after dropping the duplicated the shape of our data
data.shape
```

```python
# info about the dataframe like numbers of non-null values and datatypes of the columns
data.info()
```

```python
# all the rows having missing values in it
missing = data[data.isnull().any(axis=1)]
missing
```

```python
# count of the missing values in each column
data.isnull().sum()
```

```python
data.describe()
```

Filling the values in the numerical columns

```python
# all the columns in the data are numerical columns
# filling the missing values using the mean, median of the columns

data = data.apply(lambda x: x.fillna(x.median()) if(x.mean()>x.median()) else x.fillna(x.mean()))
```

```python
# checking the missing values are filled or not
data.isnull().sum()
```

```python
# calculating the summary statistics again
data.describe()
```
